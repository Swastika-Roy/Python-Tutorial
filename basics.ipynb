```python
#remove duplicates while preserving order
original_list = [1,2,2,3,4,3,5]

seen = set()
result = []
for item in original_list:
    if item not in seen:
        seen.add(item)
        result.append(item)

print(result)
```

```python
#remove duplicates while preserving order
original_list = [1,2,2,3,4,3,5]

def remove_duplicates(lst):
    seen = []
    for item in lst:
        if item not in seen:
            seen.append(item)
    return seen

result = print(remove_duplicates(original_list))
```
