```python
car = {
    "brand" : "Toyota",
    "model" : "Camry",
    "year" : 2020
}
car.update({"model" : "Corolla", "price" : 25000,"type" :"SEDAN"})
print("AFTER UPDATE " ,car)
del car["type"]
print("AFTER DEL " ,car)
#using pop() - remove and return val
model = car.pop("model")
print("poped val : ",model)
print("after pop() : ",car)
#using pop() default val
price = car.pop("price","not found")
print("popped price : ",price)
```

```python
dict3 = {"a" : 1, "b" : 2}
dict4 = {"b" : 3, " d" : 4}
merged = {**dict3,**dict4} #merging version 3.5
print("using ** operator : ",merged)
```

```python
dict5 = {"a" : 1, "b" : 2}
dict6 = {"b" : 3, " d" : 4}
merged = dict5 | dict6
print("using | operator : ",merged)
```

```python
dict7 = {"a" : 1, "b" : 2}
dict8 = {"b" : 3, " c" : 4}
merged = {**dict7, **dict8}
print("with duplicates : ",merged)
```

```python
student = {
    "name" : "Swastika Roy",
    "age" : 21,
    "cgpa" : "8.7",
    "stream" : "IT"
}
for key in student:
    print(key)
print()
for key in student.keys():
  print(key)
print()
for value in  student.values():
  print(value)
print()
for key,value in student.items():
  print(f"{key} : {value}")
```

```python

student = {
  "student1" : {
      "name" : " Roy",
      "age" : 21,
      "course":{
          "Maths" : "B",
          "Physics" : "A",
          "Chemistry" : "A"
      }
  },
  "student2" : {
       "name" : " M Roy",
      "age" : 22,
      "course":{
          "Maths" : "C",
          "Physics" : "A",
          "Chemistry" : "A"
      }
  },
 "student3" : {
       "name" : " D Roy",
      "age" : 24,
      "course":{
          "Maths" : "B",
          "Physics" : "A",
          "Chemistry" : "C"
      }
  }
}

print("Student name : ",student["student1"]["name"])
print("Student age : ",student["student1"]["age"])
print("Student course : ",student["student1"]["course"]["Maths"])
print()
for key,value in student.items():
  for key,value in value.items():
    print(f"{key} : {value}")
  print()
```

```python
nums = [1,2,3,4,5]
sq = {num: num**2 for num in nums}
print(sq)
cubes = {num : num**3 for num in range(1,6)}
print(cubes)
words = {"apple","banana","cherry"}
word_len = {word : len(word) for word in words}
print(word_len)
```

```python
keys = ["name","age","city"]
vals = ["alice",25,"New York"]
person = {keys[i] : vals[i] for i in range(len(keys))}
print(person)
```

```python
nums = range(1,11)
result = {num: num**2 for num in nums if num % 2 == 0}
print(result)
```

```python
words = ["hi","hello","hey","goodbye","yes"]
word_len = {word : len(word) for word in words if len(word) > 3}
print(word_len)

student = {"john":85,"emma":92,"alice":78,"bob":95}
passed_students = {name : marks for name,marks in student.items() if marks >= 90}
print(passed_students)
```
